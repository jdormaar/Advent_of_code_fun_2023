The engineer explains that an engine part seems to be missing from the engine, but nobody can figure out which one. If you can __add up all the part numbers__ in the engine schematic, it should be easy to work out which part is missing.

The engine schematic (your puzzle input) consists of a visual representation of the engine. There are lots of numbers and symbols you don't really understand, but apparently __any number adjacent to a symbol__, even diagonally, is a "part number" and should be included in your sum. (Periods (.) do not count as a symbol.)

Here is an example engine schematic:
```
467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598..
```

In this schematic, two numbers are not part numbers because they are not adjacent to a symbol: `114` (top right) and `58` (middle right). Every other number is adjacent to a symbol and so __is__ a part number; their sum is `4361`.

Of course, the actual engine schematic is much larger.

````python
def find_adjacent_numbers(schematic, row, col):
    adjacent_numbers = set()
    for i in range(max(0, row - 1), min(row + 2, len(schematic))):
        for j in range(max(0, col - 1), min(col + 2, len(schematic[0]))):
            if schematic[i][j].isdigit():
                adjacent_numbers.add(int(schematic[i][j]))
    return adjacent_numbers

def sum_of_part_numbers(schematic):
    symbols = {'#', '$', '%', '&', '*', '+', '-', '/', '=', '@'}
    part_numbers = set()

    for i in range(len(schematic)):
        for j in range(len(schematic[0])):
            if schematic[i][j] in symbols:
                part_numbers.update(find_adjacent_numbers(schematic, i, j))

    return sum(part_numbers)

# Read the schematic from your file
with open('./data/Day3.csv') as file:
    engine_schematic = [line.strip() for line in file]

# Calculate the sum of part numbers
result = sum_of_part_numbers(engine_schematic)
print("Sum of all part numbers:", result)
````
